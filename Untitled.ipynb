```python
# funkcija koja ucitava video i deli ga na frejmove
def iseciVideo(broj):
    #ucitaj video
    video = cv2.VideoCapture('video-'+str(broj)+'.avi')
    #uzima prvi frejm i javlja da li je uspo to da uradi u uspesno
    uspesno, frejm = video.read()
    slike = []
    #vrti se u vajlu dokle god je uspesno na true prekida se kada dodje do zadnje frejma u uspesno vraca false
    while uspesno:
        #ubacujem frejm po frejm u niz slike
        slike.append(frejm)
        #secem sledeci frejm i vracam da li je u speo da ga isece
        uspesno, frejm = video.read()
    return slike
```

```python
#vrsim eroziju i dilaciju da ispravim sliku
def eroDil(slika):
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    slikaRet = cv2.erode(slika, kernel, iterations=1)
    slikaRet = cv2.dilate(slikaRet, kernel, iterations=1)
    return slikaRet

def iviceHough(slika):
    #pronaliz ivice na slici
    plava = cv2.Canny(slika, 250, 250)
    plava = cv2.dilate(plava, (13,13), iterations=1)
    #spaja linije na osnovu pronadjenih ivica
    lines = cv2.HoughLinesP(plava, 1, np.pi / 180, 50, None, 180, 50) 
    return lines
```

```python
def pronadjiPlavuLiniju(slika):
    #brisem crvenu u zelenu boju
    plava = slika.copy()
    plava[:, :, 1] = 0
    plava[:, :, 2] = 0
    
    #vrsim eroziju i dilaciju da ispravim sliku
    plava = eroDil(plava)
    #pronalazi ivice i spaja ih linijom
    lines = iviceHough(plava)
    
    #izbor tacaka za pravu aritmeticka sredina
    (x,y,x1,y1) = lines[0][0]
    x1K = x
    y1K = y
    x2K = x1
    y2K = y1
    
    for k in range(1,len(lines)):
        
        (x,y,x1,y1) = lines[k][0]
        x1K += x
        x2K += x1
        y1K += y
        y2K += y1
        
    x1, y1, x2, y2 = int(x1K/len(lines)),int(y1K/len(lines)),int(x2K/len(lines)),int(y2K/len(lines))

    cv2.line(plava, (x1, y1), (x2, y2), (255,0,0), 8)

    return x1, y1, x2, y2
```

```python
#funkcija koja vraca koordinate zelene linije
def pronadjiZelenuLiniju(slika):
    #brisemo crvenu i plavu boju
    zelena = slika.copy()
    zelena[:, :, 0] = 0
    zelena[:, :, 2] = 0
    
    #vrsim eroziju i dilaciju da ispravim sliku
    zelena = eroDil(zelena)
    
     #pronalazi ivice i spaja ih linijom
    lines = iviceHough(zelena)
     
    #izbor tacaka za pravu aritmeticka sredina
    (x,y,x1,y1) = lines[0][0]
    x1K = x
    y1K = y
    x2K = x1
    y2K = y1
    
    for k in range(1,len(lines)):
        
        (x,y,x1,y1) = lines[k][0]
        x1K += x
        x2K += x1
        
        y1K += y
        y2K += y1
        
    x1, y1, x2, y2 = int(x1K/len(lines)),int(y1K/len(lines)),int(x2K/len(lines)),int(y2K/len(lines))
 
    cv2.line(zelena, (x1, y1), (x2, y2), (255,156,120), 8)
   
    return x1, y1, x2, y2
```

```python
#pronalazi brojeve na slici izdvaja ih...
def iseciBrojeve(original, binarna):
    #pronalazi mi konture na osnovu binarne slike
    slika, konture, hijerarhija = cv2.findContours(binarna.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    sortirani_regioni = []
    regioni = []
    kordinate = []
    #prolazi kros svaku konturu
    for i in range(0, len(konture)):
        kontura = konture[i]
        #uzima koridnate konture
        x,y,w,h = cv2.boundingRect(kontura)
        #uzima povrsinu konture
        area = cv2.contourArea(kontura)
        #uzimamo samo konture koji ispunjavaju sledece uslove
        if(h >= 15 and h <= 25) or (w > 10 and h >= 14) and (hijerarhija[0][i][3] == -1) and area<800:
            #isecamo sliku
            region = binarna[y:y+h+1,x:x+w+1]
            #pakujemo sliku u 28x28 i ubacujemo u niz
            regioni.append([cv2.resize(region,(28,28), interpolation = cv2.INTER_NEAREST), (x,y,w,h)])
            cv2.rectangle(original,(x-3,y-3),(x+w+3,y+h+3),(0,255,0),2)
            #pakujemo kordinate svake konture da bi na frontu imali sve cetri tacke koje odredjuju konturu
            kordinate.append([x,y,x+w,y+h])
    #sortiramo kordinate u smeru x ose
    for i in range(0,len(kordinate)):
        for j in range(i+1,len(kordinate)):
            if kordinate[i][0] > kordinate[j][0]:
                temp =  kordinate[i]
                kordinate[i] = kordinate[j]
                kordinate[j] = temp
    #takodje sortiramo i slike po x osi
    regioni = sorted(regioni, key=lambda item: item[1][0])
    sortirani_regioni = sortirani_regioni = [region[0] for region in regioni]
    
    return original, sortirani_regioni, kordinate
```

```python
#povlaci liniju u vidu niza tacaka izmedju dve zadate tacke
def povuci_liniju(xP, yP, xK, yK):
    
        tacke_linije = []
        xR = abs(xK - xP)
        yR = abs(yK - yP)
        x, y = xP, yP
        if xP>xK:
            sx = -1
        else:
            sx = 1
        if yP>yK:
            sy = -1
        else:
            sy = 1
        if xR > yR:
            err = xR / 2.0
            while x != xK:
                tacke_linije.append((x, y))
                err -= yR
                if err < 0:
                    y += sy
                    err += xR
                x += sx
        else:
            err = yR / 2.0
            while y != yK:
                tacke_linije.append((x, y))
                err -= xK
                if err < 0:
                    x += sx
                    err += yK
                y += sy
        tacke_linije.append((x, y))
        return tacke_linije
```

```python
# funkcija sa neta prilagodjena mojim potrebama sa dodatnim stvarima za pracenje kordinata i maksimalnim odstupanjem
from scipy.spatial import distance as dist
from collections import OrderedDict
import numpy as np

class Pracenje():
    def __init__(self, maxBrojTrenutakaNevidljiv = 60):
        #inicijalizacija promenjivih koje klasa poseduje
        self.objectID = 0
        self.objects = OrderedDict()
        self.nestali = OrderedDict()
        self.kordinate = OrderedDict()
        #maksimalan broj trenutaka u kojima broj moze biti nevidljiv
        self.maxBrojTrenutakaNevidljiv = maxBrojTrenutakaNevidljiv

    def register(self, objekat, kordinate):
        # ubacivanje sledeceg pronadjenog broja sa narednim slobodnim ID-jem
        self.objects[self.objectID] = objekat
        self.nestali[self.objectID] = 0
        self.kordinate[self.objectID] = kordinate
        self.objectID += 1

    def deregister(self, objectID):
        # brisanje broja koji je nestao na odredjeni broj trenutaka specificiranih poljem nevidljivi
        del self.objects[objectID]
        del self.nestali[objectID]
        del self.kordinate[objectID]

    def azuriraj(self, rects):
        if len(rects) == 0:
            # ako u trenutnom frejmu nema brojeva onda ih sve stavimo u nestale
            for objectID in self.nestali.keys():
                self.nestali[objectID] += 1

                # ako se desi da su nestali duze nestali od maxBrojTrenutakaNevidljiv izbacujemo ih
                if self.nestali[objectID] > self.maxBrojTrenutakaNevidljiv:
                    self.deregister(objectID)

            return (self.objects, self.kordinate)

        # pravimo skladiste za trenutne brojeve na frejmu
        trenutni = np.zeros((len(rects), 2), dtype="int")
        # prolazimo kroz sve trenutne brojeve na frejmu

        for temp in range(0,len(rects)):
            #trazim centar
            (startX, startY, endX, endY) = rects[temp]
            centarX = (startX + endX) / 2
            centarY = (startY + endY) / 2
            trenutni[temp] = (int(centarX),int(centarY))
            
        # ako trenutno ne pratimo ni jedan objekat(broj) potrebno je izvrsiti dodavanje brojeva
        if len(self.objects) == 0:
            for i in range(0, len(trenutni)):
                self.register(trenutni[i], rects[i])

        # ako imamo objekte onda je potrebno izvrsiti pracenje istih
        else:
            # pokupimo sve id-ijeve i kordinate svakog broja
            objectIDs = list(self.objects.keys())
            objectPracenje = list(self.objects.values())

            # racunamo distancu izmedju svakog postojeceg i objekta na trenutnom frejmu da bi pronasli 
            # objekte koji su vec bili na prethodnom frejmu
            rastojanje = dist.cdist(np.array(objectPracenje), trenutni)
            
            rows = rastojanje.min(axis=1).argsort()
            cols = rastojanje.argmin(axis=1)[rows]
            usedRows = set()
            usedCols = set()

            for (row, col) in zip(rows, cols):
                if row in usedRows or col in usedCols:
                    continue

                objectID = objectIDs[row]
                self.objects[objectID] = trenutni[col]
                self.nestali[objectID] = 0

                self.kordinate[objectID] = rects[col]

                usedRows.add(row)
                usedCols.add(col)

            unusedRows = set(range(0, rastojanje.shape[0])).difference(usedRows)
            unusedCols = set(range(0, rastojanje.shape[1])).difference(usedCols)

            if rastojanje.shape[0] >= rastojanje.shape[1]:

                for row in unusedRows:
                    
                    objectID = objectIDs[row]
                    self.nestali[objectID] += 1

                    if self.nestali[objectID] > self.maxBrojTrenutakaNevidljiv:
                        self.deregister(objectID)
            
            else:
                for col in unusedCols:
                    self.register(trenutni[col], rects[col])

        #
        return (self.objects, self.kordinate)
```

```python
def sacuvaj(sumaNiz):
    izlazniFajl = open('out.txt', 'w')
    izlazniFajl.write("RA21/2015 Dejan Predojevic" + '\n' + "file" + '\t' + "sum" + '\n')
    for i in range(0, len(sumaNiz)):
        izlazniFajl.write("video-" + str(i) + ".avi" + '\t' + str(sumaNiz[i]) + '\n')
    izlazniFajl.close()
    
def nadjiIndeks(x,y,kord):
    #trazimo index broja o kome se radi
    for temp in (range(0, len(kord))):
        if kord[temp][0] == x:
            if kord[temp][1] == y:
                return temp
    return -1;
```
